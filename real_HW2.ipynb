```python
# build training_set, test_set
training_set, test_set = sales.random_split(.8,seed=0)
# show selected features
my_features = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','zipcode']
advanced_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode','condition', 'grade', 'waterfront', 
                     'view', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated','lat', 'long', 'sqft_living15', 'sqft_lot15' ]
```

```python
my_features_model = tc.linear_regression.create(training_set,target='price',features=my_features, validation_set=None)
```

```python
ad_features_model = tc.linear_regression.create(training_set,target='price',features=advanced_features, validation_set=None)
```

```python
print("The RMSE of the model with advanced features lower by ${:.0f}".format(my_features_model.evaluate(test_set)['rmse']-ad_features_model.evaluate(test_set)['rmse']))
```
